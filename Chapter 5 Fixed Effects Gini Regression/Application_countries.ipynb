### Import Gini regressions for panels

```python
from Gini_panel import GiniFixedEffects
import pandas as pd
import numpy as np
%matplotlib inline
```

### Data: wages characteristics

```python
import linearmodels
from linearmodels.datasets import wage_panel
data = wage_panel.load()
data = data.set_index(["nr", "year"])
print(wage_panel.DESCR)
print(data.head())
```

```python
data = wage_panel.load()
X = data[["expersq", "union", "married", "hours"]]   
y = data.lwage
model = GiniFixedEffects()
model.fit(data, "nr", y, X, method='Gini', cov = 'instrument')
model.summary()
```

```python
model.fit(data, 'nr', y, X, method='Gini', estimator = 'parametric', cov = 'instrument')
model.summary()
```

```python
model.fit(data, 'nr', y, X, method='Gini', estimator = 'parametric', centering='between', cov = 'instrument')
model.summary()
```

```python
model.fit(data, 'nr', y, X, method='Gini', sampling=2)
model.summary()
```

```python
model.fit(data, 'nr', y, X, method='OLS')
model.summary()
```

### OLS panels with linearmodels 

```python
from linearmodels.panel import PanelOLS
import statsmodels.api as sm
data = wage_panel.load()
data = data.set_index(["nr", "year"])
X = data[["expersq", "union", "married", "hours"]]
X = sm.add_constant(X)
model = PanelOLS(data.lwage, X, entity_effects=True)
results = model.fit()
print(results)
```

```python
from linearmodels.panel import PooledOLS
model = PooledOLS(data.lwage, X)
results = model.fit()
print(results)
```
