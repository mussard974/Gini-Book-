```python
from GiniRegression_v2 import GiniRegression
```

###  <span style="color:red">Dataset importation</span>

```python
from sklearn.datasets import fetch_california_housing
data_california = fetch_california_housing()
X = data_california.data  
y = data_california.target 
```

```python
# Non-parametric Gini regression 
model = GiniRegression()
model.fit(y, X)
model.summary()
```

```python
# Non-parametric Gini regression 
model = GiniRegression()
model.fit(y, X, iter = 2000)
model.summary()
```

```python
#OLS
import statsmodels.api as sm
import statsmodels.regression.linear_model as sm_lm
x = sm.add_constant(X)
model_ols = sm.OLS(y,x)
results = model_ols.fit()
print(results.summary())
```

```python
# Gini parametrique 
model = GiniRegression()
model.fit(y, X, parametric_estimator= True)
model.summary()
```

```python
# Gini parametrique avec variance des paramètres par var. instrumentales
model = GiniRegression()
model.fit(y, X, parametric_estimator= True, cov='instrument') 
model.summary()
```

```python
# Gini parametrique avec variance des paramètres par var. instrumentales
model = GiniRegression()
model.fit(y, X, parametric_estimator= True, cov='asympt') 
model.summary()
```

```python
# Gini parametrique avec variance des paramètres par var. instrumentales
model = GiniRegression()
model.fit(y, X, parametric_estimator= True, cov='bandwidth iid') 
model.summary()
```

```python
# Gini parametric with cov Jackknife
model = GiniRegression()
model.fit(y, X, parametric_estimator= True, cov='Jackknife') 
model.summary()
```
